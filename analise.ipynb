```python
import csv
import seaborn as sns
import pandas as pd
```

```python
arquivo = 'resultado-00000-of-00001.csv'
df = pd.read_csv(arquivo, delimiter=';')
```

```python
df.groupby(['UF', 'ANO'])[['CHUVA','DENGUE']].mean().reset_index()
```

```python
ax = sns.relplot(x="ANO", y="DENGUE", data=df);
ax.figure.set_size_inches(12, 5)
```

```python
ax = sns.relplot(x="ANO", y="DENGUE", hue="CHUVA", data=df);
ax.figure.set_size_inches(12, 5)
```

```python
ax = sns.relplot(x="DENGUE", y="CHUVA", size="DENGUE", sizes=(15, 200), data=df);
ax.figure.set_size_inches(12, 5)
```

```python
ax = sns.relplot(x="DENGUE", y="CHUVA", size="DENGUE", style="UF", sizes=(15, 200), data=df);
ax.figure.set_size_inches(12, 5)
```
